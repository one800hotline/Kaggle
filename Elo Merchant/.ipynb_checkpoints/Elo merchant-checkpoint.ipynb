### What files do I need?
You will need, at a minimum, the train.csv and test.csv files. These contain the card_ids that we'll be using for training and prediction.

The historical_transactions.csv and new_merchant_transactions.csv files contain information about each card's transactions. historical_transactions.csv contains up to 3 months' worth of transactions for every card at any of the provided merchant_ids. new_merchant_transactions.csv contains the transactions at new merchants (merchant_ids that this particular card_id has not yet visited) over a period of two months.

merchants.csv contains aggregate information for each merchant_id represented in the data set.

#### What should I expect the data format to be?
The data is formatted as follows:

train.csv and test.csv contain card_ids and information about the card itself - the first month the card was active, etc. train.csv also contains the target.

historical_transactions.csv and new_merchant_transactions.csv are designed to be joined with train.csv, test.csv, and merchants.csv. They contain information about transactions for each card, as described above.

merchants can be joined with the transaction sets to provide additional merchant-level information.

#### What am I predicting?
You are predicting a loyalty score for each card_id represented in test.csv and sample_submission.csv.

#### File descriptions
* train.csv - the training set
* test.csv - the test set
* sample_submission.csv - a sample submission file in the correct format - contains all card_ids you are expected to predict for.
* historical_transactions.csv - up to 3 months' worth of historical transactions for each card_id
* merchants.csv - additional information about all merchants / merchant_ids in the dataset.
* new_merchant_transactions.csv - two months' worth of data for each card_id containing ALL purchases that card_id made at merchant_ids that were not visited in the historical data.

#### Data fields
* Data field descriptions are provided in Data Dictionary.xlsx.#

```python
def f_trf_train_test(exec_f, indata, meta_prefix='trn_tst_'):

    # Go!
    if exec_f:
        
        df_tmp=indata.copy()
            
        #---------------------
        # Time features
        #---------------------
        df_tmp['period_first_purch_dofw']=df_tmp['period_yyyymm_first_purch'].dt.dayofweek
        df_tmp['period_first_purch_wofy']=df_tmp['period_yyyymm_first_purch'].dt.weekofyear
        df_tmp['period_first_purch_mofy']=df_tmp['period_yyyymm_first_purch'].dt.month
        
        #--------------------------------------------
        # Group target variable for EDA purposes
        #--------------------------------------------
  
        df_tmp_train=df_tmp[df_tmp['f_train']==1]

        # bins to match histogram and capture "tail heavy" properties
        list_bin_target=[-50]+[bin_trg for bin_trg in range(-5,0,1)]+[bin_trg for bin_trg in range(0, 5, 1)]+[50]
        list_bin_target_int=[grp_val_map for grp_val_map in range(1, (len(list_bin_target)+1))]
        
        # Apply bins and create discrete groups
        df_tmp_train['grp_target']=pd.cut(df_tmp_train['target'], bins=list_bin_target, include_lowest=True)
        df_tmp_train['grp_target_int']=df_tmp_train['grp_target'].cat.codes+1
                                   
        # Merge back to common
        df_tmp=pd.concat([df_tmp_train, df_tmp[df_tmp['f_train']==0]], axis=0, sort=True)
        
        # Rename
        if meta_prefix:
            df_tmp.columns=['card_id']+[meta_prefix+col for col in df_tmp.columns if col not in (['card_id'])]
        
                
        # Push back result
        return df_tmp
            
    # No Go!
    else:
        print ("No execution of transform logic, returning indata...")
        return indata
```

```python
import time as time
tick=time.time()

#----------------------------
# Transforn trans_hist data
#----------------------------
def f_trf_hist_trans(exec_f,indata, f_trans_hist=True):
    
    if exec_f:
        
        # Safety copy
        df_temp=indata.copy()
        
        #-----------------------------
        # Transform some variables
        #-----------------------------
        
        # Map to binary categorical
        df_temp['f_auth']=np.where(df_temp['authorized_flag']=='Y'
                                  ,1
                                  ,0)
        
        # Map to binary categorical
        df_temp['f_cat1']=np.where(df_temp['category_1']=='Y'
                                   ,1
                                   ,0)
        
        # Category 2
        df_temp['grp_cat2']=df_temp['category_2']
        
        # Group variable map dict
        dict_map_cat3={'A':1
                      ,'B':2
                      ,'C':3}
        
        # Apply to create numeric
        df_temp['grp_cat3']=df_temp['category_3'].map(dict_map_cat3)
        
        # Group variable installment
        df_temp['grp_installmnt']=np.where(df_temp['installments']==-1
                                        ,-1
                                        ,np.where(df_temp['installments']==0
                                                 ,0
                                                 ,np.where(df_temp['installments']==999
                                                          ,999
                                                          ,1
                                                          )
                                                 )
                                          )
        
        # Compact form aggregator variable for illustrating time series
        df_temp['period_yr_mnth_purch']=df_temp['purchase_date'].str[0:7].str.replace('-','')
        
        #-----------------
        # Fill missing
        #----------------
            
        df_temp['grp_cat3'].fillna(1, inplace=True)

        df_temp['grp_cat2'].fillna(1, inplace=True)
        
        df_temp['merchant_id'].fillna('N/A', inplace=True)
        
        #----------------
        # Bin continious
        #----------------
        
        # bin of purchase amount to get a sense for where majority of amount end up
        list_bin_purch_amt=[-0.8,-0.7,-0.6,-0.5,-0.1]+[elem*0.01 for elem in range(0, 11,2)]+[elem for elem in range(2,10,2)]+[100]+[1000000000000]

        df_temp['grp_purch_amt']=pd.cut(df_temp['purchase_amount'], bins=list_bin_purch_amt, include_lowest=True)        

        #-----------------
        # Flag outlier
        #-----------------
        df_temp['f_outl_purch_amt']=np.where(df_temp['purchase_amount'] > 10000
                                            ,1
                                            ,0)
        
        
        # Clean up
        df_temp.drop(['authorized_flag', 'category_1','category_2', 'category_3'], axis=1, inplace=True)
        
        #------------------------------------------------------
        # Below only applied for the transaction history data, 
        # new merchant only passes through data above
        #------------------------------------------------------
        if f_trans_hist:
        
            #----------------------
            # Feature engineering
            #----------------------

            # Time derivations
            df_temp['period_purch_dow']=df_temp['period_purch_d'].dt.dayofweek
            df_temp['period_purch_week']=df_temp['period_purch_d'].dt.week
            df_temp['period_purch_mnth']=df_temp['period_purch_d'].dt.month

            # Time since purchase
            df_temp = df_temp.sort_values(by=['card_id', 'period_purch_d'], ascending=[1, 0]).reset_index(drop=True)

            # Time between purchases
            df_temp['period_purch_d_lag1']=df_temp[['card_id', 'period_purch_d']].groupby('card_id').shift(1)
            df_temp['period_diff_purch_dt']=df_temp['period_purch_d']-df_temp['period_purch_d_lag1']

            # Fill missing values since we get that from lag.
            df_temp['period_diff_purch_dt'].fillna(1, inplace=True)

            # Set to integer
            df_temp['period_diff_purch_int']=df_temp['period_diff_purch_dt'].dt.seconds/60

            # clean up
            df_temp.drop(['period_purch_d_lag1', 'period_diff_purch_dt'], axis=1, inplace=True)

            # Aggregation dictionary to create Time as Customer
            dict_feat_eng_tac={'month_lag':'count'
                              ,'period_purch_d':['min','max']
                              }


            # Time as customer
            df_temp2=cstr.f_grpby_aug(True, df_temp, ['card_id'], dict_feat_eng_tac)
            df_temp2['period_tac']=df_temp2['period_purch_d_max']-df_temp2['period_purch_d_min']
            df_temp2['period_tac_days']=df_temp2['period_tac'].dt.days

            df_temp2['period_tac_mnth']=df_temp2['period_tac']/np.timedelta64(1, 'M')


            # To main DataFrame
            df_temp=df_temp.merge(df_temp2[['card_id', 'period_tac_days','period_tac', 'period_tac_mnth']], how='inner', left_on='card_id', right_on='card_id')        

        
        # Return output
        return df_temp
        
    else: 
        print ("No exeuction of transform step for DataFrame hist_trans, ending.... ")

        return indata
    
    
# Apply transformation
df_trf_hist_trans=f_trf_hist_trans(True, df_hist_trans_sub, f_trans_hist=True)





print ("Total elapsed time is: {}s".format(round(time.time()-tick),2))
```

```python
print (f_dt_now())
tick=time.time()

def f_time_behavior_trend(exec_f, indata, list_time_lag, list_col_id_agg, list_col_period_timedelta, list_col_metric_timdelta):
    
    """
    This function looks at a reference date, and then lags the date in months backward, after this it will calculate a mean for a specified metric given available 
    rows between lagged date and reference date. Returned DataFrame has one row per id and nr of metric columns according to nr of lag time periods, e.g. if lag is 
    1m and 3m then metric_col_lg1 and metric_col_lg3m will be returned. If no rows then nan
    
    """
    
    if exec_f:
        
        df_temp=indata.copy()
   
        # Push out a series for all rows, with latest purchase date as reference
        df_temp['period_first_purch']=df_temp[list_col_id_agg+list_col_period_timedelta].groupby(list_col_id_agg).transform('first')

        print ("\n")
        
        # Itterate through the months lags we want to create columns for
        for idx_day, mnth_lag in enumerate(list_time_lag):
            
            print ("Constructing lagged vector: {}".format(mnth_lag))

            # Lag each time period given how long we want to look back and catch behavior shifts
            df_temp['period_purch'+'_lag'+str(mnth_lag)]=df_temp['period_first_purch']-pd.DateOffset(months=mnth_lag)
    

            # We create a groupby object so we only capture the rows wanted for each card_id, and time lag (1, 3, and 6 months)
            df_temp['grp_agg_timedelta'+'_lag'+str(mnth_lag)]=np.where(df_temp[list_col_period_timedelta[0]]>=df_temp['period_purch'+'_lag'+str(mnth_lag)]
                                                                       ,1
                                                                       ,0)


        # For setting names at the end
        list_period_calc_metric=['period_'+list_col_metric_timdelta[0]+'_lag'+str(mnth_lag) for mnth_lag in list_time_lag]
        
        # Aggregator variables
        list_col_agg_timedelta=['grp_agg_timedelta_lag'+str(mnth_lg) for mnth_lg in list_time_lag]

        # Aggregate to one card_id per row and nr of agg columns given nr of lags 
        df_temp_agg=pd.concat([df_temp[(df_temp[col_period_lag]==1)][list_col_id_agg+[col_period_lag]+list_col_metric_timdelta].groupby(list_col_id_agg+[col_period_lag]).mean() for col_period_lag in list_col_agg_timedelta]
                               ,axis=1).reset_index()

        # drop things we dont want
        df_temp_agg.drop(list_col_agg_timedelta, inplace=True, axis=1)

        # Rename
        df_temp_agg.columns=list_col_id_agg+list_period_calc_metric
        
        # Return DataFrame
        return df_temp_agg
        
    else:
        
        print ("No calculation of time trend variables, returning indata....")
        return indata


df_agg_trans_hist_trend_purch_amt=f_time_behavior_trend(True, df_trf_hist_trans, [1, 3, 6, 12],['card_id'], ['period_purch_d'], ['purchase_amount'])
df_agg_trans_hist_trend_installm=f_time_behavior_trend(True, df_trf_hist_trans, [1, 3, 6, 12],['card_id'], ['period_purch_d'], ['installments'])


print ("\nTotal elapsed time is: {}".format(round(time.time()-tick, 2)))

for df in ([df_agg_trans_hist_trend_purch_amt, df_agg_trans_hist_trend_installm]):
    display(df[:5])
```

```python
def f_gen_agg_dict():
    
    """
    Function that generates an automated aggregation dictionary for contruct function cstr.f_grpby_aug(). The function generates a list with statistics, one for each column name,
    and then runs them through a dictionary comprehension to map key:value pairs for estimation. This function is run pre aggregation of data 
 
    Example:
    Column list has 5 columns, and is mapped to statistics key word list_stat_v1 --> one list with 5 list_stat_v1 is created, and is then mapped to each column name in column list. 
    """
    # main agg dictionary
    dict_agg_trans_hist={}
    
    # Update 1:1 key to val mappings
    for dict_upd in list_unq_col_to_stat_map:
        dict_agg_trans_hist.update(dict_upd)
    
    # Update columns with stat list v1 
    _tmp_list_stat1=list()
    for idx in range(len(list_col_stat1)):
        _tmp_list_stat1.append(list_stat_v1)

    # Update columns with stat list v2 
    _tmp_list_stat2=list()
    for idx in range(len(list_col_stat2)):
        _tmp_list_stat2.append(list_stat_v2)

    # Update columns with stat list v3 
    _tmp_list_stat3=list()
    for idx in range(len(list_col_stat3)):
        _tmp_list_stat3.append(list_stat_v3)

    # Update columns with stat list v4 
    _tmp_list_stat4=list()
    for idx in range(len(list_col_stat4)):
        _tmp_list_stat4.append(list_stat_v4)
        
        
    # Columns to stat
    dict_lvl1={key:val for key,val in zip(list_col_stat1, _tmp_list_stat1)}
    dict_lvl2={key:val for key,val in zip(list_col_stat2, _tmp_list_stat2)}
    dict_lvl3={key:val for key,val in zip(list_col_stat3, _tmp_list_stat3)}
    dict_lvl4={key:val for key,val in zip(list_col_stat4, _tmp_list_stat4)}

    
    # Update to final dict
    for dict_upd_final in [dict_lvl1, dict_lvl2, dict_lvl3, dict_lvl4]:
        dict_agg_trans_hist.update(dict_upd_final)
    
    return dict_agg_trans_hist
```

```python
def f_find_col_map_list_statistics():

    """
    Function that itterates over a set of column names and picks out columns given find-function 
    from a reference list of names. It will then, given name of column, map a set of statistics to be calculated
    through a number of list containing key words, e.g. ['min', 'max']....
 
    Example:
    If colum name is "state", it will find "state_min", "state_max", "state_mean", etc... and store away in a list. To this a set of statistics is mapped also
    """
    
    list_hld_col_rename_all=list()

    # Over columns to find
    for idx_collvl1, col_lvl1 in enumerate(list_col_rename_mnth):

        # All columns in DataFrame
        for idx_col_find, col_find in enumerate(df_trf_hist_trans_mnth_agg):

            # If we find it we save it
            if col_find.find(col_lvl1)>=0:
                
                # Map set of list statistics 1
                if col_lvl1 in list_col_stat1:
                    list_hld_col_rename_all.append((col_find, list_stat_v1))
                    
                elif col_lvl1 in list_col_stat2:
                    list_hld_col_rename_all.append((col_find, list_stat_v2))
                    
                elif col_lvl1 in list_col_stat3:
                    list_hld_col_rename_all.append((col_find, list_stat_v3))
                    
                elif col_lvl1 in list_col_stat4:
                    list_hld_col_rename_all.append((col_find, list_stat_v4))                    
                                        
                elif col_lvl1 in list_unq_col_to_stat_map:
                    list_hld_col_rename_all.append((col_find, list_stat_unq))

    # Return list with tuple pair of column name and statistics to be calculated
    return list_hld_col_rename_all
```

```python
tick=time.time()

def f_extr_col_dups(exec_f, indata, return_list=False):

    """
    Function for pulling columns where same esimation has bee done two times, e.g. "purchase_amount_mean_mean", etc... They are pushed to a list and indata is returned with column subset.
    """
        
    # Executing function
    if exec_f:
        
        df_temp=indata.copy()
    
        # Make a metadata table
        df_temp_cols=pd.DataFrame(df_temp.columns).rename(columns={0:'col_name'})
        df_temp_cols['col_name_split_1_lst']=df_temp_cols['col_name'].str.split("_").str[-1]
        df_temp_cols['col_name_split_2_lst']=df_temp_cols['col_name'].str.split("_").str[-2]

        # Cols with last two derivations bein same, e.g. "_max_max", etc. 
        list_col_check=df_temp_cols[(df_temp_cols['col_name_split_1_lst']==df_temp_cols['col_name_split_2_lst'])]['col_name'].tolist()

        # We return list with columns
        if return_list==True:
            return list_col_check
        
        # We retun df subset on axis=1
        else:
            return df_temp[list_col_check]

    # No go! 
    else:
        print ("No execution of function, returning None")

df_temp_sub_col=f_extr_col_dups(True, df_trf_hist_trans_agg)

print ("Elapsed time is: {}s".format(round(time.time()-tick, 2)))
```

```python
def f_drop_col_unq_val(exec_f, indata, drop_level=1, return_list=False, display_df=False):

    """
    Function for looking into the nr of unique values in all columns of a DataFrame. Returns a list given drop_level, or returns a DataFrame and list of columns. 
    """
    
    if exec_f:

        # Safety copy
        df_temp=indata.copy()

        # All columns to list
        list_col_freq_val=df_temp.columns

        # list holder
        list_hld_col_val_freq=list()

        # Move through columns and check nr of unique values
        for idx_col, col_freq in enumerate(list_col_freq_val):

            # Append to common value
            set_tmp=len(df_temp[col_freq].drop_duplicates())
            list_col_tmp=col_freq
            list_df_len=len(df_temp)
            list_hld_col_val_freq.append((list_col_tmp, set_tmp, list_df_len))

        # As DataFrame
        df_col_val_unq=pd.DataFrame(list_hld_col_val_freq).rename(columns={0:'col_name'
                                                                            ,1:'nr_unq_val'
                                                                            ,2:'total_row_'})
        # List of columns to drop, given drop-level
        list_col_drop=df_col_val_unq[df_col_val_unq['nr_unq_val']<=drop_level]['col_name'].tolist()
        
        # How does it look
        if display_df:
            display(df_col_val_unq.sort_values(by='nr_unq_val')[:5])
            
        # We return both list of cols with 1 level, and DataFrame with all columns
        if return_list:
            return df_col_val_unq, list_col_drop
        
        # Only return DataFrame with unique value levels
        elif not return_list:
            return df_col_val_unq
        
    else:
        print ("No execution of function for dropping columns, returning indata")
        return indata
```

### Sample data to reduce computation time

```python
#----------------------------------------------------------
# We sample out some data to get a model up and running,
# using less information to try our way around
#----------------------------------------------------------

def f_smpl_df(exec_f, indata, axis_0_smpl=30000, axis_1_smpl=110):

    if exec_f:

        df_tmp_train=indata[indata['trn_tst_f_train']==1].copy()
        df_tmp_test=indata[indata['trn_tst_f_train']==0].copy()
        
        # Drop train/test flag since we know where we are going from here
        for df_drop in [df_tmp_train, df_tmp_test]:
            df_drop.drop(['trn_tst_f_train'], inplace=True, axis=1)
        
        #------------------- 
        # Train data
        #-------------------
        
        # Sample axis=1
        np.random.seed(1234)
        array_ids_smpl_axis_1=np.random.randint(725, size=axis_1_smpl)

        # Map dict to have index along column name
        dict_col_map_idx=dict(zip([col for col in df_tmp_train.columns]
                                 ,[idx_col for idx_col in range(0, len(df_tmp_train.columns))]
                                 )
                             )

        # fetch column names given index from dictionary
        list_col_name_smpl=[val[0] for val in dict_col_map_idx.items() if val[1] in (array_ids_smpl_axis_1)]

        # Sample axis=0
        df_tmp_train_smpl_sub=df_tmp_train.sample(axis_0_smpl)

        # Sample axis=1
        df_tmp_train_smpl_sub=df_tmp_train_smpl_sub[list_col_name_smpl+['trn_tst_target']]
        
        print ("Shape of sample train data is: {}".format(df_tmp_train_smpl_sub.shape))
        
        #----------------------------------
        # Test data (only sample columns)
        #----------------------------------
        df_tmp_test_sub=df_tmp_test[list_col_name_smpl]

        print ("Shape of sample test data is: {}".format(df_tmp_test_sub.shape))

        
        # Return result    
        return df_tmp_train_smpl_sub, df_tmp_test_sub,list_col_name_smpl
        
    else:
        print ("No sampling, returning indata.")
        return indata
        
```

### Initialize algorithms

```python
# Gradient boosinng machine parameters
lgb_params = {'num_leaves': 111,
             'min_data_in_leaf': 149, 
             'objective':'regression',
             'max_depth': 9,
             'learning_rate': 0.005,
             "boosting": "gbdt",
             "feature_fraction": 0.7522,
             "bagging_freq": 1,
             "bagging_fraction": 0.7083 ,
             "bagging_seed": 11,
             "metric": 'rmse',
             "lambda_l1": 0.2634,
             "random_state": 133,
             "verbosity": -1}

# XGBoost
xgb_params = {'eta': 0.01
              ,'max_depth': 10
              ,'subsample': 0.8
              ,'colsample_bytree': 0.8
              ,'objective': 'reg:linear'
              ,'eval_metric': 'rmse'
              ,'silent': True
              ,'nthread': 4}

# Cat gradient boosting
cat_params = {'learning_rate': 0.01,
              'depth': 12,
              'l2_leaf_reg': 10,
              'bootstrap_type': 'Bernoulli',
              #'metric_period': 500,
              'od_type': 'Iter',
              'od_wait': 50,
              'random_seed': 11,
              'allow_writing_files': False}
```

```python
# Fold settings
n_folds = 5
folds = KFold(n_splits=n_folds, shuffle=True, random_state=11)

def f_train_model(X, X_test, y, params=None, folds=folds, n_folds=n_folds, model_type='lgb',plot_feature_importance=False):

    oof = np.zeros(len(X))
    prediction = np.zeros(len(X_test))
    scores = []
    feature_importance = pd.DataFrame()
    for fold_n, (train_index, valid_index) in enumerate(folds.split(X)):
        print('Fold', fold_n, 'started at', time.ctime())
        X_train, X_valid = X.iloc[train_index], X.iloc[valid_index]
        y_train, y_valid = y.iloc[train_index], y.iloc[valid_index]
        
        if model_type == 'lgb':
            model = lgb.LGBMRegressor(**params, n_estimators = 20000, nthread = 4, n_jobs = -1)
            model.fit(X_train, y_train, 
                    eval_set=[(X_train, y_train), (X_valid, y_valid)], eval_metric='rmse',
                    verbose=1000, early_stopping_rounds=200)
            
            y_pred_valid = model.predict(X_valid)
            y_pred = model.predict(X_test, num_iteration=model.best_iteration_)
            
        if model_type == 'xgb':
            train_data = xgb.DMatrix(data=X_train, label=y_train)
            valid_data = xgb.DMatrix(data=X_valid, label=y_valid)

            watchlist = [(train_data, 'train'), (valid_data, 'valid_data')]
            model = xgb.train(dtrain=train_data, num_boost_round=20000, evals=watchlist, early_stopping_rounds=200, verbose_eval=1000, params=params)
            y_pred_valid = model.predict(xgb.DMatrix(X_valid), ntree_limit=model.best_ntree_limit)
            y_pred = model.predict(xgb.DMatrix(X_test), ntree_limit=model.best_ntree_limit)
            
        if model_type == 'rcv':
            model = RidgeCV(alphas=(0.01, 0.1, 1.0, 10.0, 100.0), scoring='neg_mean_squared_error', cv=3)
            model.fit(X_train, y_train)
            print(model.alpha_)

            y_pred_valid = model.predict(X_valid)
            score = mean_squared_error(y_valid, y_pred_valid) ** 0.5
            print(f'Fold {fold_n}. RMSE: {score:.4f}.')
            print('')
            
            y_pred = model.predict(X_test)
            
        if model_type == 'cat':
            model = CatBoostRegressor(iterations=20000,  eval_metric='RMSE', **params)
            model.fit(X_train, y_train, eval_set=(X_valid, y_valid), cat_features=[], use_best_model=True, verbose=False)

            y_pred_valid = model.predict(X_valid)
            y_pred = model.predict(X_test)
        
        oof[valid_index] = y_pred_valid.reshape(-1,)
        scores.append(mean_squared_error(y_valid, y_pred_valid) ** 0.5)
        
        prediction += y_pred    
        
        if model_type == 'lgb':
            # feature importance
            fold_importance = pd.DataFrame()
            fold_importance["feature"] = X.columns
            fold_importance["importance"] = model.feature_importances_
            fold_importance["fold"] = fold_n + 1
            feature_importance = pd.concat([feature_importance, fold_importance], axis=0)

    prediction /= n_folds
    
    print('CV mean score: {0:.4f}, std: {1:.4f}.'.format(np.mean(scores), np.std(scores)))
    
    if model_type == 'lgb':
        feature_importance["importance"] /= n_folds
        if plot_feature_importance:
            cols = feature_importance[["feature", "importance"]].groupby("feature").mean().sort_values(
                by="importance", ascending=False)[:50].index

            best_features = feature_importance.loc[feature_importance.feature.isin(cols)]

            plt.figure(figsize=(16, 12));
            sns.barplot(x="importance", y="feature", data=best_features.sort_values(by="importance", ascending=False));
            plt.title('LGB Features (avg over folds)');
        
            return oof, prediction, feature_importance
        return oof, prediction
    
    else:
        return oof, prediction
```

```python
def f_exec_lgbm_est(indata_X, y, list_col_features):

    tick=time.time()
    
    df_x_tmp=indata_X.copy()
    df_y=y.copy()
    
    folds = KFold(n_splits=5, shuffle=True, random_state=15)
    oof = np.zeros(len(df_x_tmp))
    predictions = np.zeros(len(df_x_tmp))
    start = time.time()
    feature_importance_df = pd.DataFrame()

    for fold_, (trn_idx, val_idx) in enumerate(folds.split(df_x_tmp.values, df_y.values)):
        print("fold n{}".format(fold_))
        trn_data = lgbm.Dataset(df_x_tmp.iloc[trn_idx],
                               label=df_y.iloc[trn_idx],

                              )
        val_data = lgbm.Dataset(df_x_tmp.iloc[val_idx],
                               label=df_y.iloc[val_idx],

                              )

        num_round = 10000
        clf = lgbm.train(param,
                        trn_data,
                        num_round,
                        valid_sets = [trn_data, val_data],
                        verbose_eval=1000,
                        early_stopping_rounds = 200)

        oof[val_idx] = clf.predict(df_x_tmp.iloc[val_idx], num_iteration=clf.best_iteration)

        fold_importance_df = pd.DataFrame()
        fold_importance_df["feature"] = list_col_features
        fold_importance_df["importance"] = clf.feature_importance()
        fold_importance_df["fold"] = fold_ + 1
        feature_importance_df = pd.concat([feature_importance_df, fold_importance_df], axis=0)

        predictions += clf.predict(df_x_tmp, num_iteration=clf.best_iteration) / folds.n_splits
        
    print("\nCV score: {:<8.5f}".format(mean_squared_error(oof, df_y)**0.5))
    print ("Total elapsed time is: {}s".format(round(time.time()-tick, 2)))
        
    return fold_importance_df, feature_importance_df, clf
    

```
