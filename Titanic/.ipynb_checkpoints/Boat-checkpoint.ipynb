```python
# Base functions imported in order to get methods get_params() and set_params() to return classifiers parameters as well as using the score method
from sklearn.base import BaseEstimator
from sklearn.base import ClassifierMixin

from sklearn.preprocessing import LabelEncoder
from sklearn.externals import six
from sklearn.base import clone
from sklearn.pipeline import _name_estimators
import operator


class MajorityVoteClassifier(BaseEstimator
                            ,ClassifierMixin):
    
    
    """
    A majority vote ensamble classifier
    
    Parameters:
    
    classifiers          array-like, integer, shape = [n_classifiers]: Different classifiers for the ensamble
    vote                 str, {'classlabel', 'probability'}: Default 'probability'. If 'classlabel' the prediction is based on the argmax of class labels. Else if
                                                             'probability', the argmax of the sum of the probabilities is used to predict the class label (use for calibrated classifiers)
    weights              array-like, float, shape = [n_classifiers]: Optional, default = None. If a list of integers or floats are provided, the classifiers are weighted by importance.
                                                                     Uses uniform weight if weights = None.
        
    """
    
    def __init__(self
                ,classifiers
                ,vote = 'classlabel'
                ,weights = None):
        
        self.classifiers = classifiers
        self.names_classifiers = {key: value for key, value in _name_estimators(classifiers)}
        self.vote = vote
        self.weights = weights
        
    def fit(self
           ,X
           ,y):
        
        """
        Fit classifier
        
        Parameters:
                
        X        integers: {Array-like}, shape = [n_samples, n_features]
                           Training vector, where n_samples is number of samples(rows) and n_features is number of features(columns) 
                         
        y        integers: Array-like, shape = [n_samples]
                           Target values
                           
        Returns:
        
        self      Object
        
        """        
        
        # Use LabelEncoder to ensure that class labels start with 0 which is of importance for np.argmax
        # Call in self.predict
        
        self.labelenc_ = LabelEncoder()
        self.labelenc_.fit(y)
        self.classes_ = self.labelenc_.classes_
        
        self.classifiers_ = []
        
        for clf in self.classifiers:
            fitted_clf = clone(clf).fit(X, self.labelenc_.transform(y))
            self.classifiers_.append(fitted_clf)
            
        return self
            
            
            
    def predict(self
               ,X):
        
        
        """
        
        Assign a class label through majority vote.
        
        Paramters:
        
        X               Array-like, sparse matrix, integer/float. Shape = [n_rows, n_features]: Matrix of training sample
        
        Returns:
        
        maj_vote        array-like, shape = [n_samples]: Predicted class labels from the ensamble  
        
        """
        
        if self.vote == 'probability':
            maj_vote = np.argmax(self.predict_proba(X)
                                ,axis = 1)
            
        # CLass label vote
        else:
            
            # Collects result from clf.predict calls
            predictions = np.asarray([clf.predict(X) for clf in self.classifiers_]).T
            
            maj_vote = np.apply_along_axis(lambda x: np.argmax(np.bincount(x
                                                                           ,weights = self.weights
                                                                          )
                                                              )
                                                                ,axis = 1
                                                                ,arr = predictions
                                          )
            
            maj_vote = self.labelenc_.inverse_transform(maj_vote)
            
            return maj_vote
        
    
    def predict_proba(self
                     ,X):
        
        """
        
        Predict (score) probabilities for class labels through majority vote.
        
        Parameters:
        
        X               Array-like, sparse matrix, integer/float. Shape = [n_rows, n_features]: Training vector where n_rows is number of sample and n_features is dimensionality of explanatory variables
        
        Returns:
        
        avg_proba       array-like, float, shape = [n_sample, n_classes]: Weighted average probability for each class per sample
        
        """ 
        
        probas = np.asarray([clf.predict_proba(X) for clf in self.classifiers_])
        
        avg_probas = np.average(probas
                               ,axis = 0
                               ,weights = self.weights
                               )
        
        return avg_probas
    
    def get_params(self
                  ,deep = True):
        
        """ 
        
        Get classifier parameter names for the GridSearch.
        
        """
        
        if not deep:
            return super(MajorityVoteClassifier
                        ,self).get_params(deep = False)
        
        else:
            out = self.names_classifiers.copy()
            
            for name, step in six.iteritems(self.named_classfiers):
                for key, value in six.iteritems(step.get_params(deep = True)):
                    out['%s__%s' % (name, key)] = value
                    
            return out
        
        
    
```

```python
class c_perceptron(object):
    
    """
    Perceptron classifier as described by McCullok-Pitts 1943 - A logical calculus of the ideas immanent in nervous activity. 
    
    Parameters:
    
    eta          numeric(float): Learning rate (between 0 and 1)
    n_iter       numeric(int): Passes over the training data
    
    Attributes:
    
    w_           1d-array: Weights after fitting
    errors_      list: Number of missclassifications after each epoch (pass)
    
    """
    
    def __init__ (self
                 ,eta = 0.001
                 ,n_iter = 25):
        
        self.eta = eta
        self.n_iter = n_iter
        
    def fit(self
            ,X
            ,y):
        
        """
        Fit parameters to model.
        
        Parameters:
        
        X        integers: {Array-like}, shape = [n_samples, n_features]
                           Training vector, where n_samples is number of samples(rows) and n_features is number of features(columns) 
                         
        y        integers: Array-like, shape = [n_samples]
                           Target values
                           
        Returns:
        
        self      Object
        
        """
        
        # Initialize a weight vector with shape according to number of features + intercept
        self.w_ = np.zeros(1 + X.shape[1])
        self.errors_ = list()
        
        # Itterations according to main parameters - passes over training data
        for _ in range(self.n_iter):
            errors = 0
            
            # Itteration over all rows, and all pairs of target and feature values, for selected column(s)
            for Xi, target in zip(X, y):
                
                # Weight update: If (target - prediction) = 0 then no update, else weight is pushed in direction of error (up if positive, down if negative)
                update = self.eta * (target - self.predict(Xi))
                
                # Featue weights 1 + n
                self.w_[1 : ] += update * Xi
                
                # Intercept
                self.w_[0] += update
                
                # Errors, if any, else 0
                errors += int(update != 0.0)
                
#                 print ("Itteration %s" % _) 
#                 print ("Weight for features: %s" % self.w_[1:])
#                 print ("Weight for intercept: %s" % self.w_[0])
#                 print ("Errors is: %s" % errors)
                
            # Append errors and return result
            self.errors_.append(errors)
            
        return self
         
    def net_input(self, X):
        
        """
        Calculates the net input, i.e. the matrix dot product between the weights and the input feature matrix
        """
        
        return np.dot(X, self.w_[1 : ] + self.w_[0])
    
    def predict(self, X):
        
        """
        Returns class labels after unit step function, if net input is greates than 0 we "fire", else we not
        """
        
        return np.where(self.net_input(X) >= 0
                       ,1
                       ,-1)
        
```

```python
class c_trf_load(BaseEstimator
                 ,TransformerMixin):
    
    def __init__(self, path, feat_eng = True):
        
        # Transform main features
        self.feat_eng = feat_eng
        
        # path to read csv --> DataFrame
        self.path = path
                
    def fit(self):
        
        # Do nothing
        return self
    
    def transform(self):
        
        #-------------------------------
        # Transform = True
        #-------------------------------
        if self.feat_eng:
            
            # Read data
            self.data = pd.read_csv(self.path
                                   ,index_col = False)
            
            # Set all column names to lower case
            self.data.columns = [i.lower() for i in self.data.columns]

            # Re-name "survived" to "target"
            self.data = self.data.rename(columns= {'survived' : 'target'
                                                   ,'passengerid' : 'id'})
            

            # Set sex to binary
            self.data.loc[:, 'i_gender'] = np.where(self.data.sex == 'male'
                                           ,1
                                           ,0)

            self.data.drop(labels = 'sex'
                     ,axis = 1
                     ,inplace = True)


            #---------------------
            # Cabin feature
            #---------------------
            
            # Fill cabin with common occuring value - possibly change later
            self.data.cabin.fillna('C'
                             ,inplace = True)

            
            self.data['cabin_main'] = self.data.cabin.apply(lambda i: re.findall(r"\D+", i)).str.get(0).str.strip()
            
            self.data.cabin_main = self.data.cabin_main.replace({'F G' : 'F', 'F E' : 'F'})
            
            self.replace_cabin  = dict(zip(list(self.data.cabin_main.drop_duplicates())
                                           ,[i for i in range(1, 1 + len(self.data.cabin_main.drop_duplicates()))]
                                          )
                                      )
            
            self.data['i_cabin_main'] = self.data.cabin_main.map(self.replace_cabin)
            
            self.data['i_cabin_main'].replace(0
                                              ,inplace = True)


            # Count number of cabins for a passenger
            self.data['cabin_cnt'] = self.data['cabin'].apply(lambda i: i.count(' '))
            self.data['i_cabin_cnt'] = np.where(self.data.cabin_cnt == 0
                                              ,1
                                              ,self.data.cabin_cnt + 1)



            # Max number of cabins is four for a given passenger --> We create empy columns to hold information and set them to nan
            for self.idx in range(1, 5):

                self.data['cabin_' + str(self.idx) + '_str'] = np.nan
                self.data['cabin_' + str(self.idx) + '_num'] = np.nan


            # Handle if we only have one cabin per passenger, being majority of rows
            self.data.cabin_1_str = self.data[self.data.cabin_cnt == 1]['cabin'].str.extract('([A-Z])'
                                                                              ,expand=True
                                                                             )

            self.data.cabin_1_num = self.data[self.data.cabin_cnt == 1]['cabin'].str.extract('([0-9]\w{0,})'
                                                                              ,expand=True
                                                                             )


            # Go through each row with passengers > 1 cabins
            for self.idx, self.val in enumerate(self.data[self.data.cabin_cnt > 1].index.values):

                # Loop over number of cabins for a given passanger
                for self.idx_2 in range(0, (self.data.loc[self.val, 'cabin_cnt'])):

                    self.idx_2_inc1 = self.idx_2 + 1

                    # Split out all cabins into a list 
                    self.hld_cabin = self.data.loc[self.val, 'cabin'].split()[self.idx_2]

                    # For cabin i, take out the letter and number on given cabin i
                    self.data.loc[self.val, 'cabin_' + str(self.idx_2_inc1) + '_str'] = "".join([i for i in self.hld_cabin if i.isalpha()])
                    self.data.loc[self.val, 'cabin_' + str(self.idx_2_inc1) + '_num'] = "".join([i for i in self.hld_cabin if i.isdigit()])
            
            #---------------------
            # Ticket type feature
            #---------------------
            
            
            self.data['ticket'] = self.data['ticket'].str.replace('/', ' ').str.replace('.', '')

            self.data['ticket_letter'] = self.data.ticket.apply(lambda i: re.findall(r"\D+", i)).str.get(0).str.strip()

            self.data['ticket_letter_grp'] = np.where(self.data.ticket_letter.isin(list(pd.DataFrame(self.data['ticket_letter'].value_counts()).query('ticket_letter >= 5').index))
                                                ,self.data.ticket_letter
                                                ,'OTHER')

            self.data['ticket_num'] = self.data['ticket'].replace(self.data[self.data.ticket_letter.notnull()]['ticket_letter'], '', regex = True, ).str.strip().str.replace('.', '').str.replace(' ', '')

            self.data['ticket_num2'] = self.data['ticket_num'].apply(lambda i: str(i)[0:2])

            self.data.loc[self.data.ticket_num2 == '', ['ticket_num2']] = '-99'

            self.data['ticket_num2_grp'] = np.where(self.data['ticket_num2'].isin(list(pd.DataFrame(self.data['ticket_num2'].value_counts()).query('ticket_num2 >= 4').index))
                                                    ,self.data.ticket_num2
                                                    ,'-99')     
            
            self.data['ticket_letter'] = self.data.ticket.str.replace(' ', '_')

            self.data['ticket_letter_grp'] = self.data.ticket_letter_grp.str.replace(' ', '_')
            
            self.replace_title  = dict(zip(list(self.data.ticket_letter_grp.drop_duplicates())
                                           ,[i for i in range(1, 1 + len(self.data.ticket_letter_grp.drop_duplicates()))]
                                          )
                                      )
            
            self.data['i_ticket_letter_grp'] = self.data.ticket_letter_grp.map(self.replace_title)
            
            self.data.i_ticket_letter_grp.fillna(0
                                                ,inplace = True)
            
            
            
            
            #-------------------------
            # Family feature
            #-------------------------

            self.data['i_family_size'] = self.data['sibsp'] + self.data['parch'] + 1
            
            
            #-------------------------
            # title feature
            #-------------------------
            
            # Function for searching through title
            def get_title(name):
                
                title_srch = re.search(' ([A-Za-z]+)\.', name)
                
                if title_srch:
            
                    return title_srch.group(1)
            
                return ''
                
            # Apply function
            self.data['title'] = self.data.name.apply(lambda i: get_title(i))
            
            # Clean title some more            
            self.data['title'] = np.where(self.data.title.isin(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'])
                                         ,'Rare'
                                         ,self.data.title)

            self.data['title'] = self.data['title'].str.replace('Mlle', 'Miss').str.replace('Ms', 'Miss').str.replace('Mme', 'Mrs')
            
            self.replace_title  = dict(zip(list(self.data.title.drop_duplicates())
                                           ,[i for i in range(1, 1 + len(self.data.title.drop_duplicates()))]
                                          )
                                      )
            
            self.data['i_title'] = self.data.title.map(self.replace_title)
            
#             self.data.i_title.fillna(len(self.data.title.drop_duplicates())
#                                     ,inplace = True)
                
            self.data.i_title.fillna(0
                                    ,inplace = True)
            
            
            #-------------------------
            # traveling alone feature
            #-------------------------
            
            self.data['i_f_isalone'] = np.where(self.data.i_family_size == 1
                                             ,1
                                             ,0)

            
            #-------------------------
            # Embarked
            #-------------------------
            
            self.data.embarked.fillna('S'
                                     ,inplace = True)
            
            
            self.replace_embarked  = dict(zip(list(self.data.embarked.drop_duplicates())
                                           ,[i for i in range(1, 1 + len(self.data.embarked.drop_duplicates()))]
                                          )
                                      )
            
            self.data['i_embarked'] = self.data.embarked.map(self.replace_embarked)
            

            #-------------------------
            # parch + sibsp
            #-------------------------
            
            self.data.rename(columns = {'parch' : 'i_parch'
                                        ,'sibsp' : 'i_sibsp'}
                            ,inplace = True)
            
            
            #-------------------------
            # Age feature <-- Bin it
            #-------------------------
            
            self.tmp_bin = cstr.c_binning(self.data)
            
            self.data = self.tmp_bin.binning('age'
                                            ,10
                                            ,0
                                            ,10
                                            ,999
                                            ,-999

                                            ,False)
            
            
            self.replace_age = dict(zip(list(self.data.grp_age.drop_duplicates())
                                           ,[i for i in range(1, 1 + len(self.data.grp_age.drop_duplicates()))]
                                          )
                                      )
            
            self.data['i_grp_age'] = self.data.grp_age.map(self.replace_age)
            
            
            #-------------------------
            # Fare feature <--- Bin it
            #-------------------------
            
            self.tmp_bin = cstr.c_binning(self.data)
            
            self.data = self.tmp_bin.binning('fare'
                                              ,5
                                              ,0
                                              ,20
                                              ,999
                                              ,-999

                                              ,False)
            
            self.replace_fare = dict(zip(list(self.data.grp_fare.drop_duplicates())
                                           ,[i for i in range(1, 1 + len(self.data.grp_fare.drop_duplicates()))]
                                          )
                                      )
            
            self.data['i_grp_fare'] = self.data.grp_fare.map(self.replace_fare)
            
            
            #-------------------------
            # Returned transformed data
            #-------------------------
            
            return self.data            
            
        
        #-------------------------------
        # Transform = False
        #-------------------------------
        else:
            
            # Read data
            self.data = pd.read_csv(self.path
                                   ,index_col = False)

            # Set all column names to lower case
            self.data.columns = [i.lower() for i in self.data.columns]

            # Re-name "survived" to "target"
            self.data = self.data.rename(columns= {'survived' : 'target'
                                                   ,'passengerid' : 'id'
                                                   ,'sex' : 'gender'
                                                  })
            
            return self.data
        
```

### Fill missing values:

* Filling with most common value for discrete values
* Mean values for continious

```python
class c_trf_fillna(BaseEstimator
                   ,TransformerMixin):

    def __init__(self
                ,df
                ,impute = True
                ,f_impute = True
                ,dict_imp = dict_imp
                 ):
        
        # df to be imputed
        self.df = df
        
        # Impute or not
        self.impute = impute
            
        # Flag all imputed variables
        self.f_impute = f_impute
        
        # Control dictionary on how to impute missing values
        self.dict_imp = dict_imp
            
            
    def fit(self):
        
        # Do nothnig
        return self
    
    def transform(self):
        
        
        # Impute
        if self.impute:
            
            # Flag imputed rows to let algo know we have been tampering with it
            if self.f_impute:
        
                # Move through control dictionary and apply impute operator specified
                for self.var, self.operator in zip(self.dict_imp.keys(), self.dict_imp.values()):
                    
                    if self.operator == 'cat_max':

                        # Flag
                        self.df['f_' + self.var] = np.where(self.df[self.var].isnull()
                                                               ,1
                                                               ,0)

                        # Impute max from pandas Series index
                        self.df[self.var].fillna(self.df[self.var].value_counts().idxmax()
                                            ,inplace = True)


                    elif self.operator == 'mean':

                        # Flag
                        self.df['f_' + self.var] = np.where(self.df[self.var].isnull()
                                                               ,1
                                                               ,0)
                        
                        # Fill with mean value on continous variable
                        self.df[self.var].fillna(self.df[self.var].mean()
                                            ,inplace = True)


                    elif self.operator == 'n/a':

                        # Flag
                        self.df['f_' + self.var] = np.where(self.df[self.var].isnull()
                                                           ,1
                                                           ,0)
                        # Fill with constant
                        self.df[self.var].fillna('n/a'
                                                    ,inplace = True)

                return self.df                        
                        
                
            # No flag on imputed, just apply operator
            else:

                # Move through control dictionary and apply impute operator specified
                for self.var, self.operator in zip(self.dict_imp.keys(), self.dict_imp.values()):
                    
                    # Max value for categorical variable
                    if self.operator == 'cat_max':

                        self.df[self.var].fillna(self.df[self.var].value_counts().idxmax()
                                                ,inplace = True)


                    elif self.operator == 'mean':

                        self.df[self.var].fillna(self.df[self.var].mean()
                                                ,inplace = True)


                    elif self.operator == 'n/a':

                        self.df[self.var].fillna('n/a'
                                                ,inplace = True)
                
                return self.df
        
        # No impute
        else:
            
            return self.df
```

### Important considerations when finalizing machine learning models.

* ***Python Version.*** Take note of the python version. You almost certainly require the same major (and maybe minor) version of Python used to serialize the model when you later load it and deserialize it.
* ***Library Versions.*** The version of all major libraries used in your machine learning project almost certainly need to be the same when deserializing a saved model. This is not limited to the version of NumPy and the version of scikit-learn.
* ***Manual Serialization.*** You might like to manually output the parameters of your learned model so that you can use them directly in scikit-learn or another platform in the future. Often the algorithms used by machine learning algorithms to make predictions are a lot simpler than those used to learn the parameters can may be easy to implement in custom code that you have control over.

```python
# Empy holder for each perceptron modell
hld_perc = dict()
n_models = 5

for idx, rnd_seed in enumerate((np.random.randint(1, 100, (n_models, )))):
    
    print ("Initialization seed for perceptron classifier is: %s \n" % rnd_seed)
    
    clf_perc = perceptron.Perceptron(random_state = rnd_seed)

    pipe_perc = Pipeline([('scl', StandardScaler())
                           ,('clf' , clf_perc)])

    clf__alpha = np.linspace(0.001, 1, 10)
    clf__n_iter = [i for i in range(50, 550, 50)]
    clf__penalty = ['l1', 'l2']

    param_grd1 = {'clf__alpha' : clf__alpha
                 ,'clf__n_iter' : clf__n_iter
                 ,'clf__penalty' : clf__penalty     # If added, lowers the accuracy
                }

    gs_perc = GridSearchCV(estimator = pipe_perc
                           ,param_grid = param_grd1
                           ,scoring = 'accuracy'
                           ,cv = 10
                           ,n_jobs = 1)



    gs_perc_fit = gs_perc.fit(X_train
                            ,y_train)

    gs_optimal_mdl_perc = gs_perc_fit.best_estimator_


    y_train_pred = gs_optimal_mdl_perc.predict(X_train)
    y_test_pred = gs_optimal_mdl_perc.predict(X_test)

    cv_score =  cross_val_score(gs_optimal_mdl_perc
                               ,X_train
                               ,y_train
                               ,scoring = 'accuracy'
                               ,cv = 5)

    hld_perc.update(zip(['mdl_' + str(rnd_seed)], [gs_optimal_mdl_perc]))

    print ("|-------------------------------------------------------------------------------------|")
    print ("               Executing model construction for itteration %s and seed %s" % ((idx + 1), rnd_seed))  
    print ("|-------------------------------------------------------------------------------------|")
    print ("Best grid search parameters: %s" % gs_perc.best_params_)
    print ("\n")
    print("Accuracy: %0.2f (+/- %0.2f)" % (cv_score.mean(), cv_score.std() * 2))
    print ("\n")

    print ("Classification report: \n")
    print (classification_report(y_train, y_train_pred))
```

## Stacking

### Tools:

* Sklearn wrap class
* out-of-fold function tool


In order to more easily automate the process of constructing classifiers, a wrapper class i constructed for the Sklearn classifiers avilable, and their most important methods. 

To support creating oof predictions on the input data, to be fed from the level 0 classifiers into the level 1 meta classifier, a out-of-fold function is created. In order to to allow the classifiers to see all data, i.e. when creating clf on full data and potentially overit the algo, oof is applied.

```python
print ("Train shape: %s \n Test shape: %s" % (X_train.shape, X_test.shape))
```

```python
# Important parameters to be used when creating stacking architecture and oof predictions
n_train = X_train.shape[0]
n_test = X_test.shape[0]
seed = 123
n_folds = 5 # for oof predictions
kf = KFold(n_train
           ,n_folds = n_folds
           ,random_state = seed)

# Extent Sklearn classifiers (train, fit and predict methods)
class c_SklearnWrap(object):
    
    def __init__(self
                ,clf
                ,seed = 123
                ,params = None):
        
        params['random_state'] = seed
        self.clf = clf(**params)
        
    def train(self
             ,X_train
             ,y_train):
    
        self.clf.fit(X_train
                    ,y_train)
        
    def predict(self
               ,X):
    
        return self.clf.predict(X)
    
    def fit(self
           ,X_train
           ,y_train):
    
        return self.clf.fit(X_train
                           ,y_train)
```

```python
def f_get_oof(exec_f
              ,clf
              ,X_train
              ,y_train
              ,X_test):
    
    """
    Docstring for function get_oof (get out of fold) predictions when applying stacking architecture. Note that the number of models trained in total is a product of the number of folds used
    for creating the oof predictions. E.g., if 5 folds is used 
    
    
    """
    
    
    if exec_f:
        
        # For holding index values, size (n, d) according to adherent input data (train, test and n_fold))
        oof_train = np.zeros((n_train, ))
        oof_test = np.zeros((n_test, ))
        oof_test_skf = np.empty((n_folds, n_test))
        
        # NOTE: Test index here is the index set getting oof prediction on training data, i.e. if folds is 5 we train model on train_idx k-1 (4) data points 
        # and predict on k-4 folds (1) being the test_idx
        for idx, (train_idx, test_idx) in enumerate(kf):
            
            # Create data given current fold, and loop itteration
            X_trn = X_train[train_idx]
            y_trn = y_train[train_idx]
            X_tst = X_train[test_idx]
            
            # Train given classifier on above fold-data
#             clf.train(X_trn
#                      ,y_trn)
            
            # Predict both on the training data and test data. The test data gets k predictions per model. If default k is 5, and models are 5, then a total of 25 predictions are made
            oof_train[test_idx] = clf.predict(X_tst)
            oof_test_skf[idx, :] = clf.predict(X_test)
                    
        # The test data gets k (5 as default) predictions per model, hence the mean
        oof_test[:] = oof_test_skf.mean(axis = 0)
        
        # Push back the prediction results
        return oof_train.reshape(-1, 1), oof_test.reshape(-1, 1)
        
    else:
        
        print ("No exeuction of function, ending...")
        
        
```

```python
# Function for loading models in pickel format into an object from given path
def f_fetch_models(path):

    hld_loaded_mdls = list()
    hld_mdls_names = list()
    
    files = os.listdir(path)
    
    model_files = [i for i in files if i.find('.pkl') > 0]

    # Push out what we get
    for idx, mdl in enumerate(model_files):

        idx += 1

        print ("Model # %s is model: %s" % (idx, mdl))
        
        hld_loaded_mdls.append(joblib.load(path + mdl))
        hld_mdls_names.append(mdl.replace('.pkl', ''))

    # Return object
    return dict(zip(hld_mdls_names
                    ,hld_loaded_mdls
                   )
               )

# Save into list dictionary object
hld_models = f_fetch_models('C:\\Anaconda3\\Examples\\kaggle\\Titanic models\\')
```
